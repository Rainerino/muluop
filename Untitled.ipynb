```python
import threading

import zenoh

print("init")
zenoh.init_log_from_env_or("error")

print("Scouting...")
scout = zenoh.scout(what="peer|router")
threading.Timer(1.0, lambda: scout.stop()).start()

for hello in scout:
    print(hello)
    
print("Done")
```
